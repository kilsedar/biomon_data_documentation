### Interpretation

The circular material use rate$^{1}$, also called Circularity rate, measures the share of material recovered and fed back into the economy - thus saving extraction of primary raw materials - in overall material use. It is defined as the ratio of the circular use of materials (U) to the overall material use (M), in percentage. 

A higher circularity rate value indicates that more secondary materials substitute for primary raw materials thus reducing the environmental impacts of extracting primary material. 

### More infomation

* The overall material use (M) 

    M is measured by summing up the aggregate domestic material consumption (DMC) and the circular use of materials (U) ($M = DMC + U$)
    
    
* Domestic material Consumption 

    DMC is an indicator derived from economy-wide material flow accounts (EW-MFA) collected under [Regulation (EU) 691/2011](http://eur-lex.europa.eu/legal-content/EN/TXT/?uri=CELEX:02011R0691-20140616). DMC represents the overall amount of primary raw materials used by an economy. It equals domestic extraction plus imports minus exports.  
    
* The circular use of materials (U) 

    U is approximated by the amount of waste recycled in domestic recovery plants ($RCV_{R}$), minus imported waste destined for recovery ($IMP_{w}$), plus exported waste destined for recovery abroad ($EXP_{w}$) ($U = RCV_{R} - IMP_{w} + EXP_{w}$)
    
    Waste recycled in domestic recovery plants is approximated by the amount of hazardous and non-hazardous waste treated in recovery plants excluding amounts used for energy recovery and backfilling (recovery operations R2 to R11 as defined in the Waste Framework Directive [75/442/EEC](http://eur-lex.europa.eu/legal-content/EN/TXT/?qid=1504263102941&uri=CELEX:31975L0442)). This amount of treated waste is corrected by imports and exports of waste. 
    
    The circular material use rate is part of the “SDG 12 Ensure sustainable consumption and production patterns”: https://unstats.un.org/sdgs/report/2019/goal-12/

### More information
The circular material use rate is calculated using three European statistics: 

1. Treatment of waste by waste category, hazardousness and waste operations ([env_wastrt](https://ec.europa.eu/eurostat/web/products-datasets/-/env_wastrt)) collected on the basis of the Waste Statistics Regulation (EC) [No 2150/2002](https://eur-lex.europa.eu/legal-content/EN/ALL/?uri=CELEX:32002R2150), 
2. economy-wide material flow accounts ([env_ac_mfa](https://ec.europa.eu/eurostat/web/products-datasets/-/env_ac_mfa)), and
3. [international trade in goods statistics (comext)](https://ec.europa.eu/eurostat/web/international-trade-in-goods/overview). 

The amounts of waste treated in domestic recovery operations (1) are corrected by net exports of waste destined for recycling sourced from Comext database (3). 

A Eurostat paper of 2018 presents the reference methodology for the [calculation method of the circular material use rate](https://ec.europa.eu/eurostat/web/products-manuals-and-guidelines/-/KS-GQ-18-013?inheritRedirect=true&redirect=%2Feurostat%2Fpublications%2Fmanuals-and-guidelines).
